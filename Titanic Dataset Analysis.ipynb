### Checking null values

```python
titanic_df.isnull().sum()
```

```python
titanic_df.isnull().sum()/len(titanic_df)*100
```

```python
titanic_df.groupby("Sex")["Age"].median()
```

```python
first_condition = (titanic_df["Sex"]=="female")
second_condition = (titanic_df["Age"].isnull())
female_age_nan_df = titanic_df[first_condition & second_condition]

third_condition = (titanic_df["Sex"]=="male")
fourth_condition = (titanic_df["Age"].isnull())
male_age_nan_df = titanic_df[third_condition & fourth_condition]

female_list = list(female_age_nan_df.index)
male_list = list(male_age_nan_df.index)

titanic_df.loc[female_list,"Age"] = 27
titanic_df.loc[male_list,"Age"] = 29
```

```python
titanic_df["Embarked"].mode()
```

```python
titanic_df["Embarked"].fillna("S",inplace=True)
```

```python
titanic_df.drop(columns=["Name","Ticket","Cabin"],inplace=True)
titanic_df.head()
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.shape
```

#### Insights:

* It is observed that there are some null values in the *Age*, *Cabin* and *Embarked* columns.
* The null values in the *Age* and *Embarked* columns are filled using fillna() function.
* In *Age* column, the null values for female passengers is filled by the median age of the female passengers. And null values for the male passengers is filled by the median age of the male passengers.
* In *Embarked* column, the null values are filled by the mode of the column.
* It is observed that percentage of missing values in the *Cabin* column is more than 77. Hence it has been discarded.
* Also *Name* and *Ticket* columns does not provide any useful analytical information. Hence they are removed.

### Handling outliers

#### *Age* column

```python
titanic_df = titanic_df[titanic_df["Age"] <= 70]
```

```python
titanic_df[titanic_df["Age"] > 70]
```

#### *Fare* column

```python
titanic_df = titanic_df[titanic_df["Fare"] <= 100]
```

```python
titanic_df
```

```python
titanic_df.reset_index(drop=True,inplace=True)
titanic_df
```

#### Insights:

The outliers in the *Age* and *Fare* columns have been handled by applying filtering condition.

### Handling categorical values

```python
titanic_df = pd.get_dummies(titanic_df,columns=["Sex","Embarked"])
titanic_df
```

```python
titanic_df.shape
```

#### Insights:

The categorical columns have been handled using **one hot encoding**.

### Handling numerical values

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
titanic_df[["Age"]] = scaler.fit_transform(titanic_df[["Age"]])
titanic_df
```

```python
from sklearn.preprocessing import MinMaxScaler

minmaxscaler = MinMaxScaler()
titanic_df[["Fare"]] = minmaxscaler.fit_transform(titanic_df[["Fare"]])
titanic_df
```

#### Insights:

* The *Age* column has been handled using **StandardScaler** as the column is not skewed.
* *Fare* column has been handled using **MinMaxScaler** as the column is skewed.
